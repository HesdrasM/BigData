# Dados Ausentes
    • Número de dados ausentes em cada coluna da tabela.

```python
conceito_enade.isnull().sum()
```

```python
import numpy as np 
from sklearn.impute import SimpleImputer
```

     Copiando base "conceito_enade" para realizar transformações:

```python
base_conceito_enade = conceito_enade.copy()
base_conceito_enade.shape
base_conceito_enade.head(50)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='constant')
```

## Substituindo valores ausentes por valor fixo.

```python
imputer.fit(base_conceito_enade.iloc[:, 2:7])
base_conceito_enade.sample(3)
```

```python
base_conceito_enade.iloc[:, 2:7] = imputer.transform(base_conceito_enade.iloc[:, 2:7])
base_conceito_enade.head()
```

```python
base_conceito_enade1 = conceito_enade.copy()
base_conceito_enade1['Conceito Enade (Faixa)'] = base_conceito_enade1['Conceito Enade (Faixa)'].fillna(0)
base_conceito_enade1.head()
```

```python
values = {'Conceito Enade (Faixa)': 3}
base_conceito_enade1.fillna(value=values)
```
